```python
from ripe.atlas.cousteau import AtlasRequest
import sys
sys.path.append("../")
from Backend.RAICAT.utils import (
    
    check_dns_measurements,
    dns_between_dates,
    date_range
)
countries_data=dns_between_dates("2020-01-01", "2020-06-30")
```

```python
dates = date_range("2020-01-01", "2020-06-30")
```

```python
import matplotlib.pyplot as plt
#! Dasha don't forget to pip install matplotlib in your venv

def plot_graph(data):
    # Get a sorted list of all unique countries in the data
    countries = sorted(list(set().union(*data)))

    # Create a line graph for each country
    for country in ["USA", "FRA", "IRL", "ITA", "CAN", "ISR"]:
        # Get the values for this country at each day
        values = [day.get(country, None) for day in data]

        # Plot the data
        plt.plot(dates, values, label=country)

    # Add labels and a legend
    plt.xlabel("Day")
    plt.ylabel("Value")
    plt.legend()

    plt.xticks(ticks=range(0, len(data), 10), rotation=45, fontsize=6)
    plt.tight_layout()
    # Display the plot
    plt.show()


plot_graph(countries_data)
# Day numbers reperesent index of dates array (see previous cell)
```
